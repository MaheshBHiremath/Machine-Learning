```python
#Predict the maximum temperature
import pandas as pd
df=pd.read_csv('weather.csv')
df
```

```python
df['MinTemp'].groupby(df['MinTemp']).first()
df['MinTemp'].isna().sum()
df['MinTemp']=df['MinTemp'].astype(float)
```

```python
df['MeanTemp'].groupby(df['MeanTemp']).first()
df['MeanTemp'].isna().sum()
df['MeanTemp']=df['MeanTemp'].astype(float)
```

```python
df['MaxTemp'].groupby(df['MaxTemp']).first()
df['MaxTemp'].isna().sum()
df['MaxTemp']=df['MaxTemp'].astype(float)
```

```python
x=df[['MinTemp','MeanTemp']]
y=df['MaxTemp']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(x,y,test_size=0.25)
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(X_train,Y_train)
pred=lr.predict(X_test)
print(pred)
```

```python
score=lr.score(x,y)
print(score)
```
