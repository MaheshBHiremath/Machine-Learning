```python
import pandas as pd
df = pd.read_csv("kdata.csv")
x=df[['x','y']]
y=df['class']
print(x)
```

```python
import numpy as np
import matplotlib.pyplot as plt
#plt.scatter(y,np.arange(len(y)))
color=[]
for i in y:
    if i=='positive':
        color.append('g')
    else:
        color.append('r')
color
plt.scatter(x.x,x.y,color=color)

    
```

```python
import seaborn as sns
sns.scatterplot(x='x',y='y',data=df,hue='class')
```

```python
from sklearn.neighbors import KNeighborsClassifier
k=KNeighborsClassifier(n_neighbors=5)
k.fit(x,y)
pred=k.predict([[7,2]])
print(pred)
```

```python
from sklearn.neighbors import KNeighborsClassifier
k=KNeighborsClassifier(n_neighbors=5,weights='distance') # to choose the values based on k and distance
k.fit(x,y)
pred=k.predict([[7,2]])
print(pred)
```

```python
#KNN on shirts 
df1=pd.read_csv("shirt.csv")
x=df1.drop(columns='T-ShirtSize')
y=df1['T-ShirtSize']
from sklearn.neighbors import KNeighborsClassifier
#from sklearn.model_selection import train_test_split
#X_train,X_test,Y_train,Y_test=train_test_split(x,y,test_size=0.25)
k1=KNeighborsClassifier(n_neighbors=3)
k1.fit(x,y)
ypred=k1.predict([[180,80]])
print(ypred)
```
