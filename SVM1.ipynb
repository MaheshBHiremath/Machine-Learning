```python
import pandas as pd
df = pd.read_csv("Svm_data.csv")
x=df[['x','y']]
y=df['class']
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(x,y,test_size=0.25)

from sklearn.svm import SVC

#clf = SVC(kernel='linear',C=0.001)#for linear line
#clf = SVC(kernel='poly',degree=8,C=0.001) # default degree = 3 # for poly line or curve
clf = SVC(kernel='rbf',gamma='auto',C=0.001)
clf.fit(X_train,Y_train)
pred=clf.predict(X_test)
print(pred)
```

```python
import matplotlib.pyplot as plt
plt.scatter(x['x'],x['y'],c=y)
```

```python
clf=SVC(kernel='linear',C=0.001)
clf.fit(x,y)
ax = plt.gca() # info of graph
# limits of axes
xlim=ax.get_xlim()
ylim=ax.get_ylim()
#print(xlim)
#print(ylim)
import numpy as np
xx=np.linspace(xlim[0],xlim[1],200)
yy=np.linspace(ylim[0],ylim[1],200)
print(xx.shape)
YY,XX=np.meshgrid(yy,xx)
print(XX.shape)
xy=np.vstack([XX.ravel(),YY.ravel()]).T  # ravel is used to get array as list
print(xy.shape)
z=clf.decision_function(xy).reshape(XX.shape)
plt.scatter(x['x'],x['y'],c=y)
ax.contour(XX, YY, z, colors='r', levels=[-1,0,1], alpha=0.5, linestyles=['--', '-', '--'])
b=ax.scatter(clf.support_vectors_[:, 0], clf.support_vectors_[:, 1], s=100, linewidth=1, facecolors='r')
plt.legend([b ], ["Support Vector"])
```

```python
#SVM on diabetes
import pandas as pd
df1=pd.read_csv("diabetes.csv")
x=df1[['Pregnancies','Glucose','BloodPressure','SkinThickness','Insulin','BMI','DiabetesPedigreeFunction','Age']]
y=df1["Outcome"]
#print(x)
df1['BMI']=df1['BMI'].astype(int)
df1['DiabetesPedigreeFunction']=df1['DiabetesPedigreeFunction'].astype(int)
#print(df1['DiabetesPedigreeFunction'].dtype)
x=df1[['Pregnancies','Glucose','BloodPressure','SkinThickness','Insulin','BMI','DiabetesPedigreeFunction','Age']]
#print(x)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(x,y,test_size=0.25)
from sklearn.svm import SVC
sv=SVC(kernel='linear',C=0.001)
sv.fit(X_train,Y_train)
predt=sv.predict(X_test)
print(predt)
```

```python
from sklearn.metrics import accuracy_score
ac=accuracy_score(predt,Y_test)
print(ac)
```
